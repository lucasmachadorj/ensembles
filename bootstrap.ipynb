```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

% matplotlib inline
```

```python
B = 200
N = 20
X = np.random.randn(N) # N normal random samples

print "sample mean of X: {}".format(X.mean())
```

```python
individual_estimates = np.empty(B)
for b in range(B):
    sample = np.random.choice(X, size=N)
    individual_estimates[b] = sample.mean()
```

```python
bootstrap_mean = individual_estimates.mean()
bootstrap_std = individual_estimates.std()

print "bootstrap mean of X: {}".format(bootstrap_mean)
print "bootstrap standard deviation of X: {}".format(bootstrap_std)
```

```python
lower = bootstrap_mean + norm.ppf(0.025)*bootstrap_std
upper = bootstrap_std + norm.ppf(0.975)*bootstrap_std

lower2 = X.mean() + norm.ppf(0.025)*X.std()/np.sqrt(N)
upper2 = X.mean() + norm.ppf(0.975)*X.std()/np.sqrt(N)
```

```python
plt.figure(figsize=(12,6))
plt.hist(individual_estimates, bins=20)
plt.axvline(x=lower, linestyle='--', color='g', label='lower bound for 95% CI bootstrap')
plt.axvline(x=upper, linestyle='--', color='g', label='upper bound for 95% CI bootstrap')
plt.axvline(x=lower2, linestyle='--', color='r', label='lower bound for 95% CI')
plt.axvline(x=upper2, linestyle='--', color='r', label='lower bound for 95% CI')
plt.legend()
plt.show()
```
